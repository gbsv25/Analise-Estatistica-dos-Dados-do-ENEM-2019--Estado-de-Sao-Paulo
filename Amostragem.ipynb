## **AMOSTRAGEM ESTRATIFICADA**

Foi dividido uma população em grupos (estratos) e selecione amostras fornecidas de cada grupo.

Nesse código, a estratificação foi feita com base na variável RACA, garantindo que cada grupo estivesse representado na amostra.

```
enem_campinas['RACA'].value_counts()
```

**Para quantidade pequena de grupos**

```
branca = enem_campinas.loc[enem_campinas.RACA == 'branca'].sample(frac=0.15)
branca.shape
```

```
parda = enem_campinas.loc[enem_campinas.RACA == 'parda'].sample(frac=0.15)
parda.shape
```

```
preta = enem_campinas.loc[enem_campinas.RACA == 'preta'].sample(frac=0.15)
preta.shape
```

```
amarela = enem_campinas.loc[enem_campinas.RACA == 'amarela'].sample(frac=0.15)
amarela.shape
```

```
nd = enem_campinas.loc[enem_campinas.RACA == 'nao_declarado'].sample(frac=0.15)
nd.shape
```

```
indigena = enem_campinas.loc[enem_campinas.RACA == 'indígena'].sample(frac=0.15)
indigena.shape
```

```
indigena
```

```
amos_estra = pd.concat([branca, parda, preta, amarela, nd, indigena])
```

```
amos_estra
```

**Grandes quantidades de grupos**

```
from sklearn.model_selection import train_test_split
```

```
X_treino, X_teste, y_treino, amos_estra = train_test_split(enem_campinas.drop('RACA',axis=1),enem_campinas['RACA'],
                                                    stratify=enem_campinas['RACA'], test_size=0.20)
```

```
amos_estra.shape
```

```
amos_estra.value_counts()
```

```
amos_estra = amos_estra.index
```

```
amos_estra
```

```
amos_estra_camp = enem_campinas.loc[amos_estra,:]
```

```
amos_estra_camp
```
